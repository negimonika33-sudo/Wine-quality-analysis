```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from fontTools.varLib.plot import plotDocument

#print("imported")

data = pd.read_csv("Wine Quality Dataset.csv")
data.head()

print(data.head())
print(data.describe())
print(data.info())
print(data.index)
print(data.columns)
print(data.shape)
```

```python
plt.figure(figsize=(11,6))

sns.histplot(data=data,x='fixed acidity',color='red',edgecolor ='orange',alpha=0.5,bins=5,)
plt.title('histogram of fixed acidity')
plt.xlabel('fixed acidity')
plt.ylabel('count')
plt.savefig('histogram.png')
plt.show()

# normaly distributed and max value between 6 and 8
```

```python
print(data['fixed acidity'].mean())
print(data['fixed acidity'].median())
```

```python

# create a histogram for a fixed acidity and also show mean and median
plt.figure(figsize = (11,6))

sns.histplot(data = data,x='fixed acidity',color='red',edgecolor ='orange',alpha = 0.5,bins = 5,)
plt.title('histogram of fixed acidity with mean ,median')
plt.xlabel('fixed acidity')
plt.ylabel('count')
plt.vlines(data['fixed acidity'].mean(),ymin = 0,ymax = 4000,color='blue',label='mean')
plt.vlines(data['fixed acidity'].median(), ymin = 0, ymax = 4000,color='black',label='median')
plt.legend()
plt.savefig('histogram.png')
plt.show()
```

```python

#2  create a histogram for a volatile acidity feature

plt.figure(figsize=(11,6))

sns.histplot(data=data,x='volatile acidity',color='red',edgecolor ='orange',alpha=0.5,bins=5,)
plt.title('histogram of volatile acidity')
plt.xlabel('volatile acidity')
plt.ylabel('count')
plt.savefig('histogram.png')
plt.show()
```

```python

#  create a histogram for a volatile acidity and also show mean and median

plt.figure(figsize = (11,6))

sns.histplot(data = data,x='volatile acidity',color='red',edgecolor ='orange',alpha = 0.5,bins = 5,)
plt.title('histogram of volatile acidity with mean,median')
plt.xlabel('volatile acidity')
plt.ylabel('count')
plt.vlines(data['volatile acidity'].mean(),ymin = 0,ymax = 4000,color='blue',label='mean')
plt.vlines(data['volatile acidity'].median(), ymin = 0, ymax = 4000,color='black',label='median')
plt.legend()
```

```python


plt.figure(figsize = (11,6))

sns.distplot(data['volatile acidity'],color='red'),
plt.title('histogram of volatile acidity')
plt.xlabel('volatile acidity')
plt.ylabel('count')
plt.savefig('histogram.png')
plt.show()

# OBSERVATIONS
# normal distribution most part of it
# bell curve due to its shape
# we use distplot when we will det a skewed output,to make it normal distribution
```

```python


#  Calculating skewness of volatile acidity

print(data['volatile acidity'].skew())

# OBSERVATION
# S>1 HENCE POSITIVE
```

```python


# calculation mean and median

print(data['volatile acidity'].mean)
print(data['volatile acidity'].median)
```

```python


#3 citric acid

plt.figure(figsize=(11,6))

sns.histplot(data=data,x='citric acid',color='red',edgecolor ='orange',alpha=0.5,bins=5,)
plt.title('histogram of citric acidity')
plt.xlabel('citric acidity')
plt.ylabel('count')
plt.savefig('histogram.png')
plt.show()
```

```python


#  create a histogram for a citric acid and also show mean and median

plt.figure(figsize = (11,6))

sns.histplot(data = data,x='citric acid',color='red',edgecolor ='orange',alpha = 0.5,bins = 5,)
plt.title('histogram of citric acidity with mean,median')
plt.xlabel('citric acidity')
plt.ylabel('count')
plt.vlines(data['citric acid'].mean(),ymin = 0,ymax = 4000,color='blue',label='mean')
plt.vlines(data['citric acid'].median(), ymin = 0, ymax = 4000,color='black',label='median')
plt.legend()
plt.show()

# same procedure  we can follow for other numerical column for mean,median (parameters of central tendency)
```

```python

# create a count plot for 'quality' feature.
plt.figure(figsize=(11,6))

sns.countplot(x='quality', data=data)
plt.title("count plot of quality")
plt.xlabel('quality')
plt.ylabel('count')
plt.savefig('count.png')
plt.show()

# observations-6 is the higest count quality and 9 is negligible.
```

```python


print(data['quality'].value_counts())

# calculate mode of the 'quality' feature
print(data['quality'].value_counts().index[0])

# observations
# we saw distribution of various features in the data set using appropriate plotDocument()
# calculated central tendency like mean ,mode,median for the various features
# mean and median of all the above features were similar, so we can choose the mean in these cases
# the mode of the quality feature can be chosen as a representative value
```
